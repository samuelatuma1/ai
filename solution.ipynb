```python
def children(depth, finds):
    parent = (2 ** depth) - 1
    tree = {}
    
    
    
    def grow_tree(parent, height):
        left_child = parent - 2**(height)
        right_child = parent - 1
        tree[left_child] = parent
        tree[right_child] = parent
        
        
        for find in finds:
            if find in tree.keys():
                finds.remove(find)
        
        if height == 1 or len(finds) == 0:
            return
        
        grow_tree(left_child, height - 1)
        grow_tree(right_child, height -1)
        
    grow_tree(parent, depth - 1)
    return tree
```

```python
def solution(h, q):
    import copy
    node = children(h, copy.deepcopy(q))
    ans = []
    
    for i in q:
        
        try:
            
            ans.append(node[i])
        except:
            ans.append(-1)
    return ans

solution(50, [7, 8, 11])
```
